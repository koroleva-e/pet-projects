### 3.2 Retention Rate

#### 3.2.1 Зададим функцию для получения профилей пользователей `get_profiles()`

```python
def get_profiles(df):
    '''
    сортируем сессии по ID пользователя и дате события
    группируем по ID и находим первые значения event_time и source
    столбец с временем первого события назовём first_ts
    '''
    profiles = (
        df.sort_values(by=['user_id', 'event_time'])
        .groupby('user_id')
        .agg({'event_time': 'first', 'source': 'first'})
        .rename(columns={'event_time': 'first_ts'})
        .reset_index()  # возвращаем user_id из индекса
    )
    
    '''
    определяем дату первого посещения
    и первый день месяца, в который это посещение произошло
    эти данные понадобятся для когортного анализа
    '''    
    profiles['dt'] = profiles['first_ts'].dt.date # дата первого посещения
    profiles['month'] = profiles['first_ts'].astype('datetime64[M]') # первый день месяца, в который это посещение произошло
    
    # добавим цд
    target_event = (df.query('target_event==True')
                .groupby('user_id', as_index=False)
                .agg({'event_name':'count'})
                .rename(columns={'event_name':'target_event_count'})
                .sort_values(by='target_event_count', ascending=False))  

    # добавляем признак пользователей c цд
    profiles['target_event'] = profiles['user_id'].isin(target_event['user_id'].unique())
    
    return profiles
```

Поставим горизонт на **15 дней**, чтобы пользователи успели "отжить" свой срок. Он кажется достаточным, чтобы успеть выполнить цд

```python
analysis_horizon = 15 # горизонт анализа
```

#### 3.2.4 Зададим функцию для расчета удержания `get_retention()`

```python
def get_retention(
    profiles, df, observation_date, horizon_days, dimensions=[], ignore_horizon=False
):
    
        # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        df[['user_id', 'event_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days

    # рассчитываем удержание
    # новый вариант с dimensions
    result_grouped = result_raw.pivot_table(
        index=dimensions,  # заменили dt
        columns='lifetime',
        values='user_id',
        aggfunc='nunique',
    )
    cohort_sizes = (
        result_raw.groupby(dimensions)  # заменили dt
        .agg({'user_id': 'nunique'})
        .rename(columns={'user_id': 'cohort_size'})
    )
    result_grouped = cohort_sizes.merge(
        result_grouped, on=dimensions, how='left'  # заменили dt
    ).fillna(0)
    result_grouped = result_grouped.div(result_grouped['cohort_size'], axis=0)

    # исключаем все лайфтаймы, превышающие горизонт анализа
    result_grouped = result_grouped[
        ['cohort_size'] + list(range(horizon_days))
    ]

    # восстанавливаем столбец с размерами когорт
    result_grouped['cohort_size'] = cohort_sizes

    # возвращаем таблицу удержания и сырые данные
    return result_raw, result_grouped
```

#### 3.2.5 Посчитаем retention через вызов функции

```python
retention_raw, retention = get_retention(
    profiles, df, datetime(2019, 11, 4).date(), 15, dimensions=['dt']
)  # передаём dt в качестве параметра dimensions

retention
```

### 3.3 Время, проведённое в приложении

#### 3.3.1 Отсортируем датасет

#### 3.3.2 Создадим столбец `pre_time` с временем предыдущего действия пользователя

#### 3.3.4 Вычислим среднее время между событиями для определения таймаута

**Таймаут** - это время, по истечении которого сессия пользователя считается неактивной. Если пользователь не взаимодействует с системой в течение установленного таймаута, время до следующего события должно быть приостановлено.

Для расчета таймаута найдем медианную разницу времени между событиями из `session_time`.

```python
# вытащим уникальных пользователей, количество событий и сумму времени межсобытий
timeout_df = (df.groupby('user_id', as_index=False)
             .agg({'event_name':'count', 'session_time':'sum'})
             .rename(columns={'event_name':'events_count', 'session_time':'session_time_sum'}))
timeout_df
```

```python
# добавим столбец среднего времени одного межсобытия в мин. (`session_time_sum` делим на `events_count`)
timeout_df['minutes_avg'] = (timeout_df['session_time_sum'] / timeout_df['events_count']).dt.total_seconds()/60
timeout_df.reset_index()
```

```python
timeout_df['minutes_avg'].describe()
```

```python
timeout_df['events_count'].describe()
```

👀 По медиане выходит, что пользователи совершают около 9 событий с межсобытием в 4 мин (9 * 4 = 36). Поэтому для таймаута нам будет достаточно 40 минут

Среднее время пользователя в приложении лучше считать, учитывая данные сервера, где можно увидеть подгрузку трафика от юзера - будет более точный расчет. А так, зависит от степени конфиденциальности приложения. Чем меньше, тем безопаснее

#### 3.3.5 Пронумеруем сессии в `session_id`

- сравним `session_time` относительно тайм-аута в 40 минут (порог максимальной продолжительности разовой сессии)
- если интервал >40 минут, то сессии будет присваиваться следующий номер

### 3.5 Конверсия в целевое действие (цд) — просмотр контактов

####  3.5.1 Зададим функцию для расчета конверсии `get_conversion()`

```python
def get_conversion(profiles,df,observation_date,horizon_days,dimensions=[],ignore_horizon=False):

    # Шаг 1. Получить пользовательские профили и данные о покупках
    # передаём их в качестве аргументов profiles и df

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')
    

    # Шаг 2. Найти дату и время первой покупки для каждого пользователя
    first_target = (df.query('target_event==True')
                .sort_values(by=['user_id', 'event_time'])
                .groupby('user_id')
                .agg({'event_time': 'first'})
                .reset_index())


    # Шаг 3. Добавить данные о покупках в профили
    result_raw = result_raw.merge(
        first_target[['user_id', 'event_time']], on='user_id', how='left'
    )
    

    # Шаг 4. Рассчитать лайфтайм для каждой покупки
    # расчёт дней, прошедших с первой покупки
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days
    
    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):

        # Шаг 5. Построить таблицу конверсии
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )

        # Шаг 6. Посчитать сумму с накоплением для каждой строки
        result = result.fillna(0).cumsum(axis = 1)

        # Шаг 7. Вычислить размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )

        # Шаг 8. Объединить таблицы размеров когорт и конверсии
        # добавляем размеры когорт к треугольной таблице
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)

        # Шаг 9. Разделить каждую «ячейку» на размер когорты
        # делим каждую «ячейку» на соответствующий размер когорты
        result = result.div(result['cohort_size'], axis=0)

        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)
    
    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time 
```

#### 3.5.2 Посчитаем конверсию через вызов функции

## 5. Проверка гипотез

### 5.1 Группы `yandex` и `google` демонстрируют разную конверсию в просмотры контактов

- H₀ - Конверсия `yandex` в просмотры контактов = Конверсия `google` в просмотры контактов

- H₁ - Конверсия `yandex` в просмотры контактов ≠ Конверсия `google` в просмотры контактов (двусторонняя альтернативная гипотеза)

alpha = 0.05

В проверке гипотез нам поможет Z-тест пропорций, но сначала подготовим расчеты

```python
# посчитаем общее кол-во уникальных пользователях по группам
yandex_total = df.query('source == "yandex"').user_id.nunique()
google_total = df.query('source == "google"').user_id.nunique()

print(yandex_total, google_total)
```

```python
# посчитаем кол-во уникальных пользователях с цд по группам
yandex_target = df.query('source == "yandex" & event_name == "посмотрел телефон"').user_id.nunique()
google_target = df.query('source == "google" & event_name == "посмотрел телефон"').user_id.nunique()

print(yandex_target, google_target)
```

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

target = np.array([yandex_target, google_target])
total = np.array([yandex_total, google_total])

stat, pvalue = proportions_ztest(target, total, alternative='two-sided')
print('{0:0.3f}'.format(pvalue))

# условный оператор с выводом строки с ответом
if (pvalue < alpha): 
        print('Отвергаем нулевую гипотезу, между выборками есть статистически значимые различия')
else:
        print('Не получилось отвергнуть нулевую гипотезу, статистически значимых различий в выборках нет')
```

👀 Высока вероятность, что между пользователями, установившими приложение из `yandex` или `google`, нет значимой разницы. Можно сделать вывод, что источник установки на конверсию не влияет

### 5.2 Среднее время, проведенное в приложении, у групп пользователей, совершивших целевое действие (просмотр контактов) и не совершивших его - равно

- H₀ - Среднее время, проведенное в приложении, у группы с цд = Среднее время, проведенное в приложении, у группы без цд

- H₁ - Среднее время, проведенное в приложении, у группы с цд ≠ Среднее время, проведенное в приложении, у группы без цд (двусторонняя альтернативная гипотеза)

alpha = 0.05

Проверять будем через t-тест

```python
# срез target_event==True 
target_time = (df.query('target_event==True')
                  .groupby('user_id')
                  .agg({'minutes':'first'})
                  .reset_index()) 

target_time
```

```python
# срез target_event==False
total_time = (df.query('target_event==False')
                  .groupby('user_id', as_index=False)
                  .agg({'minutes':'first'})
                  .reset_index()) 

total_time
```

```python
# results = вызов метода для проверки гипотезы
results = (st.ttest_ind(target_time['minutes'], 
                        total_time['minutes'], 
                        equal_var=False, alternative='two-sided'))

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

# вывод значения p-value на экран
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, между выборками есть статистически значимые различия')
else:
    print('Не получилось отвергнуть нулевую гипотезу, статистически значимых различий в выборках нет') 
```

👀 Отвергаем нулевую гипотезу и делаем предположение, что среднее время, проведенное в приложении, у групп пользователей, совершивших целевое действие (просмотр контактов) и не совершивших его - различается
